```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("cancer_expression.csv")

data.head()
```

```python
control_cols = [i for i in data.columns if "Control" in i]
treatment_cols = [i for i in data.columns if "Treatment" in i]

print("Columns of Controls: ", control_cols)
print("Columns of Treatments: ", treatment_cols)
```

```python
gene = "Gene10"
gene_data = data[data["Gene"] == gene]

control_mean = gene_data[control_cols].mean(axis=1).values[0]
treatment_mean = gene_data[treatment_cols].mean(axis=1).values[0]

plt.bar("Control", control_mean, label="Control", color="blue")
plt.bar("Treatment", treatment_mean, label="Treatment", color="orange")

plt.ylabel("Expression Level")
plt.title(f"Mean Expression of {gene}")
plt.legend()
plt.show()

bars = plt.bar(
    ["Control", "Treatment"],
    [control_mean, treatment_mean],
    color=["blue", "orange"]
)

plt.title(f"Bar Chart: {gene} Average Expression")
plt.ylabel("Expression Level")
plt.legend(bars, ["Control", "Treatment"],
           bbox_to_anchor=(1.05, 1),
           loc='upper left')
plt.show()
```

```python
control_data = data[["Control1", "Control2"]].values.flatten()
treatment_data = data[["Treatment1", "Treatment2"]].values.flatten()

plt.figure(figsize=(10, 6))
sns.boxplot(data=[control_data, treatment_data], palette="Set3")

plt.xticks([0, 1], ["Control", "Treatment"])
plt.title(f"Box Plot for {gene}")
plt.ylabel("Expression Level")

plt.show()
```

```python
gene = "Gene2"
gene_row = data[data["Gene"] == gene].iloc[0]

gene_data = [gene_row[control_cols].values, gene_row[treatment_cols].values]

sns.violinplot(data=gene_data)

plt.xticks([0, 1], ["Control", "Treatment"])
plt.title(f"Violin Plot: {gene} Expression Distribution")
plt.ylabel("Expression Level")

plt.show()
```

```python
heatmap_data = data.set_index("Gene")
heatmap = heatmap_data.head(10)

plt.figure(figsize=(15, 8))
sns.heatmap(heatmap, cmap="coolwarm", annot=True)

plt.title("Heatmap of Gene Expression Levels")
plt.xlabel("Sample")
plt.ylabel("Gene")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ========================================
# STEP 1: LOAD YOUR DATA
# Assuming the data is saved as 'volcano_data.csv'
# If running in a script, save the provided data to this file.
# For demo, you can copy-paste the data into a file.
df = pd.read_csv('volcano_data.csv')

# Print first few rows to verify
print(df.head())

# ========================================
# STEP 2: CONFIGURE COLUMN NAMES
# Based on your dataset:
FC_COLUMN = 'EFFECTSIZE'  # Use EFFECTSIZE as proxy for log2 fold change / effect size
PVAL_COLUMN = 'P'         # P-value column
GENE_COLUMN = 'GENE'      # Gene names for optional labeling

# ========================================
# STEP 3: PREPARE DATA
# Calculate -log10(p-value)
df['-log10_p'] = -np.log10(df[PVAL_COLUMN])

# Handle p-value = 0 or very small (set max to avoid inf)
df['-log10_p'] = df['-log10_p'].replace([np.inf, -np.inf], 300).clip(upper=300)

# ========================================
# STEP 4: SET THRESHOLDS
# For genomics/GWAS data: Genome-wide significance often P < 5e-8, |effect| > 0 (or custom)
# Adjust as needed; here using P < 5e-8 and |EFFECTSIZE| > 0.05 as example
fc_threshold = 0.05   # |EFFECTSIZE| > 0.05 (adjust based on your data scale)
p_threshold = 5e-8    # P < 5e-8 (genome-wide significance)

# ========================================
# STEP 5: COLOR POINTS
# Gray for non-significant, red for significant
df['color'] = 'gray'
significant = (abs(df[FC_COLUMN]) >= fc_threshold) & (df[PVAL_COLUMN] <= p_threshold)
df.loc[significant, 'color'] = 'red'

# Optional: Separate colors for positive/negative effect
# df.loc[(df[FC_COLUMN] > fc_threshold) & (df[PVAL_COLUMN] <= p_threshold), 'color'] = 'red'  # Positive
# df.loc[(df[FC_COLUMN] < -fc_threshold) & (df[PVAL_COLUMN] <= p_threshold), 'color'] = 'blue' # Negative

# ========================================
# STEP 6: CREATE PLOT
plt.figure(figsize=(12, 8))

# Scatter plot with smaller points for large datasets
scatter = plt.scatter(df[FC_COLUMN], df['-log10_p'],
                      c=df['color'], alpha=0.5, s=10, edgecolors='none')

# Threshold lines
plt.axhline(y=-np.log10(p_threshold), color='blue', linestyle='--', linewidth=1.5, label=f'P = {p_threshold}')
plt.axvline(x=fc_threshold, color='blue', linestyle='--', linewidth=1.5)
plt.axvline(x=-fc_threshold, color='blue', linestyle='--', linewidth=1.5)

# Labels & Styling
plt.xlabel('Effect Size', fontsize=14, fontweight='bold')
plt.ylabel('-log10 (P-value)', fontsize=14, fontweight='bold')
plt.title('Volcano Plot from GWAS Data', fontsize=16, fontweight='bold', pad=20)

# Count and display significant points
sig_count = significant.sum()
plt.text(0.02, 0.98, f'Significant SNPs: {sig_count}', transform=plt.gca().transAxes,
         fontsize=12, verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

# Legend (customize based on colors used)
plt.legend(['Non-significant', 'Significant'], loc='upper right')

plt.grid(True, alpha=0.3)
plt.tight_layout()

# Show plot
plt.show()

# ========================================
# STEP 7: SAVE RESULTS (Optional)
# Save the plot and updated DataFrame
plt.savefig('volcano_plot.png', dpi=300, bbox_inches='tight')
df.to_csv('volcano_results.csv', index=False)
print(f"Plot saved as 'volcano_plot.png'! {sig_count} significant points found.")
```
